```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Load training data
train_df = pd.read_csv("D:/DataScience/Case Study/Elevate labs/Final_project/data/twitter_training.csv", header=None, names=["id", "account_name", "sentiment_type", "message"])

print("Shape of training data:", train_df.shape)
print("\nFirst few rows:\n", train_df.head())

```

```python

# Missing values
print("\nMissing values:\n", train_df.isna().sum())
print("\nPercentage missing:\n", (train_df.isna().mean() * 100).round(2))

```

```python
# Sentiment distribution
print("\nSentiment counts:\n", train_df["sentiment_type"].value_counts())

plt.figure(figsize=(6,4))
sns.countplot(x="sentiment_type", data=train_df, order=train_df["sentiment_type"].value_counts().index)
plt.title("Sentiment Distribution")
plt.show()

```

```python
# Tweets per account
print("\nTop 10 accounts with most tweets:\n", train_df["account_name"].value_counts().head(10))

plt.figure(figsize=(8,4))
train_df["account_name"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Accounts in Training Data")
plt.ylabel("Number of Tweets")
plt.show()

```

```python
# Tweet length distribution
train_df["tweet_length"] = train_df["message"].astype(str).apply(len)

plt.figure(figsize=(6,4))
sns.histplot(train_df["tweet_length"], bins=50, kde=True)
plt.title("Tweet Length Distribution")
plt.xlabel("Characters")
plt.show()

```

```python
# Sentiment vs tweet length
plt.figure(figsize=(6,4))
sns.boxplot(x="sentiment_type", y="tweet_length", data=train_df)
plt.title("Tweet Length by Sentiment")
plt.show()
```
